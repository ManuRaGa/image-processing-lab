**1. Configurar el emulador.**

El programa lo podemos utilizar como emulador de toda la gama de cámaras de visión artificial cognex o para conectar con un visor o cámara concreta, configurarla y guardar en ella el programa o tarea de reconocimiento visual realizado.

Ingresa al siguiente link

https://support.cognex.com/es-mx/InsightEmulatorKey

Crea una cuenta y sigue los pasos que hay se muestran para generar la clave del programa y poder usarlo.

Cargar las imágenes de la carpeta Spice Jar Inspection
C:\Users\Public\Documents\Cognex\In-Sight\In-Sight Explorer 5.5.0\Sample Jobs\EasyBuilder\4x\Filter Examples\Spice Jar Inspection
